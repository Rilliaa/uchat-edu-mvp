```python
# ================================================================
# 1Ô∏è‚É£ Mount Google Drive
# ================================================================
from google.colab import drive
drive.mount('/content/drive')

# ================================================================
# 2Ô∏è‚É£ Install & Import Library
# ================================================================

import pandas as pd
import numpy as np
import torch
import re
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from datasets import Dataset
import os
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    DataCollatorWithPadding,
    TrainingArguments,
    Trainer,
    pipeline
)
os.environ["WANDB_MODE"] = "disabled"

# ================================================================
# 3Ô∏è‚É£ Load Dataset
# ================================================================
data_path = "/content/drive/My Drive/Colab Notebooks/Uchat/Dataset/dataset.csv"
df = pd.read_csv(data_path)

print("Total data:", len(df))
print("Kolom:", list(df.columns))
df.head()

# ================================================================
# 4Ô∏è‚É£ Preprocessing Text
# ================================================================
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9√†-≈æ\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Tambahkan konteks role ke dalam text
df['clean_text'] = df.apply(lambda r: f"[ROLE:{r['role']}] {clean_text(r['utterence'])}", axis=1)
df['label'] = df['intent_name']

print("\n‚úÖ Contoh data setelah preprocessing dengan ROLE context:")
print(df[['role', 'utterence', 'clean_text', 'label']].head(10))

# ================================================================
# 5Ô∏è‚É£ Stratified Split agar distribusi intent seimbang
# ================================================================
train_df, test_df = train_test_split(
    df[['clean_text', 'label']],
    test_size=0.2,
    stratify=df['label'],
    random_state=42
)
print("\nData train:", len(train_df), "| Data test:", len(test_df))

# Convert ke HuggingFace Dataset
train_ds = Dataset.from_pandas(train_df)
test_ds = Dataset.from_pandas(test_df)

# Encode label ‚Üí angka
full_ds = Dataset.from_pandas(df[['clean_text', 'label']])
full_ds = full_ds.class_encode_column("label")
label_names = full_ds.features["label"].names

# Mapping manual agar konsisten dengan label encoding
label2id = {name: i for i, name in enumerate(label_names)}
id2label = {i: name for i, name in enumerate(label_names)}

train_ds = train_ds.map(lambda x: {"label": label2id[x["label"]]}, remove_columns=["label"])
test_ds = test_ds.map(lambda x: {"label": label2id[x["label"]]}, remove_columns=["label"])

print(f"Jumlah label unik: {len(label_names)}")

# ================================================================
# 6Ô∏è‚É£ Tokenization
# ================================================================
model_name = "cahya/bert-base-indonesian-522M"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_fn(example):
    return tokenizer(example["clean_text"], truncation=True, padding=False, max_length=64)

tokenized_train = train_ds.map(tokenize_fn, batched=True)
tokenized_test = test_ds.map(tokenize_fn, batched=True)
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# ================================================================
# 7Ô∏è‚É£ Load Pretrained Model
# ================================================================
num_labels = len(label_names)
model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id
)

# ================================================================
# 8Ô∏è‚É£ Training Arguments & Trainer
# ================================================================
training_args = TrainingArguments(
    output_dir="/content/drive/My Drive/Colab Notebooks/Uchat/Model/IndoBERTweet",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=4,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    greater_is_better=True
)

def compute_metrics(eval_pred):
    from evaluate import load
    acc = load("accuracy")
    f1 = load("f1")
    prec = load("precision")
    rec = load("recall")
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": acc.compute(predictions=preds, references=labels)["accuracy"],
        "precision": prec.compute(predictions=preds, references=labels, average="macro")["precision"],
        "recall": rec.compute(predictions=preds, references=labels, average="macro")["recall"],
        "f1": f1.compute(predictions=preds, references=labels, average="macro")["f1"]
    }

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# ================================================================
# 9Ô∏è‚É£ Fine-tuning Model
# ================================================================
trainer.train()

# ================================================================
# üîü Evaluasi Hasil
# ================================================================
metrics = trainer.evaluate()
print("\nüìä Evaluation Results:")
print(metrics)

# Confusion Matrix Visual
preds = np.argmax(trainer.predict(tokenized_test).predictions, axis=1)
labels = np.array(test_ds["label"])

plt.figure(figsize=(12, 10))
cm = confusion_matrix(labels, preds)
sns.heatmap(cm, cmap="Blues", xticklabels=False, yticklabels=False)
plt.title("Confusion Matrix - IndoBERTweet Intent Classification (ROLE Context)")
plt.show()

print("\nClassification Report:")
print(classification_report(labels, preds, target_names=label_names, zero_division=0))

# ================================================================
# 1Ô∏è‚É£1Ô∏è‚É£ Simpan Model dan Tokenizer
# ================================================================
save_path = "/content/drive/My Drive/Colab Notebooks/Uchat/Model/IndoBERTweet/final_model"
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)
print(f"‚úÖ Model disimpan di {save_path}")


```

```python
# ================================================================
# 1Ô∏è‚É£2Ô∏è‚É£ Tes Utterance Manual
# ================================================================
nlp = pipeline("text-classification", model=save_path, tokenizer=save_path, top_k=None)

def predict_intent(role, text):
    text_with_role = f"[ROLE:{role}] {text}"

    # nlp() mengembalikan list of lists of dicts ketika top_k=None dan input tunggal
    list_of_results = nlp(text_with_role)

    # Kemudian kita cari skor tertinggi dari daftar kamus tersebut
    top_result = max(list_of_results[0], key=lambda x: x['score']) # Pastikan mengambil dari list dalam

    # *** Perubahan di sini ***
    # Jika top_result['label'] sudah berupa nama intent (misal: 'admin_teacher_add'),
    # kita bisa langsung menggunakannya.
    intent_name = top_result['label']
    # Jika Anda masih butuh id2label mapping, pastikan label dari model cocok dengan key di id2label

    return intent_name, top_result['score']


test_utterances = [
    ("admin", "Tolong tambahkan data guru baru"),
    ("ortu", "Saya ingin tahu nilai anak saya tahun 2024"),
    ("murid", "Tampilkan grafik kehadiran saya untuk tahun ajaran 20241"),
    ("admin", "Hapus data murid yang salah")
]

for role, utt in test_utterances:
    intent, conf = predict_intent(role, utt)
    print(f"üó£Ô∏è [{role}] '{utt}' ‚Üí üéØ {intent} (confidence={conf:.2f})")
```
