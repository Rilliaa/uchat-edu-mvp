```python
# ================================================================
# 1️⃣ Mount Google Drive
# ================================================================
from google.colab import drive
drive.mount('/content/drive')

# ================================================================
# 2️⃣ Install Library (Fix untuk Python 3.12 + Transformers modern)
# ================================================================
!pip install --upgrade transformers accelerate datasets evaluate --quiet

# ================================================================
# 3️⃣ Import Library
# ================================================================
import pandas as pd
import numpy as np
import torch
import re
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from datasets import Dataset
import os

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    DataCollatorWithPadding,
    TrainingArguments,
    IntervalStrategy,
    Trainer
)

os.environ["WANDB_MODE"] = "disabled"

# ================================================================
# 4️⃣ Load Dataset
# ================================================================
data_path = "/content/drive/My Drive/Colab Notebooks/Uchat/Dataset/dataset_focus_mode.csv"
df = pd.read_csv(data_path)

print("Total data:", len(df))
print("Kolom:", list(df.columns))
df.head()

# ================================================================
# 5️⃣ Preprocessing
# ================================================================
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9à-ž\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df["clean_text"] = df.apply(
    lambda r: f"[ROLE:{r['role']}] {clean_text(r['utterence'])}", axis=1
)

df["label"] = df["intent_name"]

print("\nContoh after preprocessing:")
print(df[['role', 'utterence', 'clean_text', 'label']].head(5))

# ================================================================
# 6️⃣ Stratified Split
# ================================================================
train_df, test_df = train_test_split(
    df[["clean_text", "label"]],
    test_size=0.2,
    stratify=df["label"],
    random_state=42
)

train_ds = Dataset.from_pandas(train_df)
test_ds = Dataset.from_pandas(test_df)

# Encode label ke angka (global)
full_ds = Dataset.from_pandas(df[["clean_text", "label"]])
full_ds = full_ds.class_encode_column("label")
label_names = full_ds.features["label"].names

label2id = {name: i for i, name in enumerate(label_names)}
id2label = {i: name for i, name in enumerate(label_names)}

train_ds = train_ds.map(lambda x: {"label": label2id[x["label"]]}, remove_columns=["label"])
test_ds  = test_ds.map(lambda x: {"label": label2id[x["label"]]}, remove_columns=["label"])

print(f"Total intent: {len(label_names)}")

# ================================================================
# 7️⃣ Tokenizer
# ================================================================
model_name = "cahya/bert-base-indonesian-522M"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(example):
    return tokenizer(
        example["clean_text"],
        truncation=True,
        padding=False,
        max_length=64
    )

tokenized_train = train_ds.map(tokenize, batched=True)
tokenized_test = test_ds.map(tokenize, batched=True)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

# ================================================================
# 8️⃣ Load IndoBERT
# ================================================================
num_labels = len(label_names)

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id
)

# ================================================================
# 9️⃣ TrainingArguments (versi aman)
# ================================================================
training_args = TrainingArguments(
    output_dir="/content/drive/My Drive/Colab Notebooks/Uchat/Model/FM IndoBERTweet",

    # Evaluasi setiap epoch
    eval_strategy=IntervalStrategy.EPOCH,
    save_strategy=IntervalStrategy.EPOCH,
    logging_strategy=IntervalStrategy.STEPS,

    learning_rate=3e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=4,
    weight_decay=0.01,

    logging_steps=50,
    load_best_model_at_end=True,
)

# ================================================================
# 🔟 Metrik Evaluation
# ================================================================
from evaluate import load
acc_metric = load("accuracy")
f1_metric = load("f1")
prec_metric = load("precision")
rec_metric = load("recall")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": acc_metric.compute(predictions=preds, references=labels)["accuracy"],
        "precision": prec_metric.compute(predictions=preds, references=labels, average="macro")["precision"],
        "recall": rec_metric.compute(predictions=preds, references=labels, average="macro")["recall"],
        "f1": f1_metric.compute(predictions=preds, references=labels, average="macro")["f1"]
    }

# ================================================================
# 1️⃣1️⃣ Trainer
# ================================================================
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_test,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

# ================================================================
# 1️⃣2️⃣ Fine-tuning
# ================================================================
trainer.train()

# ================================================================
# 1️⃣3️⃣ Evaluation
# ================================================================
metrics = trainer.evaluate()
print("\nEvaluation Results:")
print(metrics)

# Confusion Matrix
preds = np.argmax(trainer.predict(tokenized_test).predictions, axis=1)
labels = np.array(test_ds["label"])

plt.figure(figsize=(12, 10))
cm = confusion_matrix(labels, preds)
sns.heatmap(cm, cmap="Blues")
plt.title("Confusion Matrix — IndoBERT Intent Classification with ROLE Context")
plt.show()

print("\nClassification Report:")
print(classification_report(labels, preds, target_names=label_names, zero_division=0))

# ================================================================
# 1️⃣4️⃣ Save Model
# ================================================================
save_path = "/content/drive/My Drive/Colab Notebooks/Uchat/Model/FM IndoBERTweet"
model.save_pretrained(save_path)
tokenizer.save_pretrained(save_path)
print(f"✅ Model disimpan di {save_path}")


```

```python
from transformers import pipeline

# Load dari local path atau HuggingFace
nlp = pipeline(
    "text-classification",
    model=save_path,
    tokenizer=save_path,
    top_k=None,           # untuk mendapatkan seluruh skor
    function_to_apply="softmax"   # memastikan skor probabilitas
)

def predict_intent(role, text):
    text_with_role = f"[ROLE:{role}] {text}"

    outputs = nlp(text_with_role)

    # Untuk input tunggal + top_k=None → outputs = [ [ {label, score}, {...} ] ]
    predictions = outputs[0]

    # Cari label dengan skor tertinggi
    best = max(predictions, key=lambda x: x['score'])

    intent_name = best['label']
    confidence = best['score']

    return intent_name, confidence


# ================================================================
# 1️⃣2️⃣ Tes Manual
# ================================================================
test_utterances = [
    ("[ROLE:guru]", "lihat jadwal saya untuk besok"),
    ("[ROLE:ortu]", "Saya ingin cek nilai anak saya"),
    ("[ROLE:murid]", "Saya ingin cek prestasi saya Rio Liando"),
    ("[ROLE:admin]", "Redirect ke halaman pengaturan dong")
]

for role, utt in test_utterances:
    intent, conf = predict_intent(role, utt)
    print(f"🗣️ {role} '{utt}' → 🎯 {intent} (confidence={conf:.2f})")
```
