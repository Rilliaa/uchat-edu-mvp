```python
# ================================================================
# 1️⃣ Mount Google Drive
# ================================================================
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import re

# --- TAMBAHKAN KODE INI ---
# Mengatur opsi tampilan Pandas agar tidak memotong output
pd.set_option('display.max_columns', None)   # Menampilkan semua kolom
pd.set_option('display.max_rows', None)      # Menampilkan semua baris (hati-hati jika data sangat besar)
pd.set_option('display.width', None)        # Mengatur lebar tampilan agar sesuai layar
pd.set_option('display.max_colwidth', None) # Menampilkan lebar kolom penuh (tidak memotong teks panjang)
# --------------------------


# ================================================================
# 3️⃣ Load Dataset
# ================================================================
data_path = "/content/drive/My Drive/Colab Notebooks/Uchat/Dataset/dataset.csv"
df = pd.read_csv(data_path)

print("Total data:", len(df))
print("Kolom:", list(df.columns))
df.head()

# ================================================================
# 4️⃣ Preprocessing Text (dengan ROLE context)
# ================================================================
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9à-ž\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Tambahkan konteks role ke dalam text
df['clean_text'] = df.apply(lambda r: f"[ROLE:{r['role']}] {clean_text(r['utterence'])}", axis=1)
df['label'] = df['intent_name']

print("\n✅ Contoh data setelah preprocessing dengan ROLE context:")
print(df[['role', 'utterence', 'clean_text', 'label']].head(10))
```
